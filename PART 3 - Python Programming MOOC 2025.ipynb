* String operations
Strings can be combined, or concatenated, with the + operator:

begin = "ex"
end = "ample"
word = begin+end
print(word)
output : example


The * operator can also be used with a string, when the other operand is an integer. The string operand is then repeated the number of times specified by the integer. For example this would work:

word = "banana"
print(word*3)
output : bananabananabanana


Using string operations together with a loop we can write a program which draws a pyramid:

n = 10 # number of layers in the pyramid
row = "*"

while n > 0:
    print(" " * n + row)
    row += "**"
    n -= 1

This prints out the following:

          *             -> n = 10, row = *
         ***            -> n = 9, row = ***  
        *****           -> n = 8, row = *****
       *******          -> n = 7, row = *******
      *********         -> n = 6, row = *********
     ***********        -> n = 5, row = ***********
    *************       -> n = 4, row = *************
   ***************      -> n = 3, row = ***************
  *****************     -> n = 2, row = *****************
 *******************    -> n = 1, row = *******************


The print command within the loop prints a line, which begins with n spaces, followed by whatever is stored in the variable row. Then two stars are added to the end of the variable row, and the value of the variable n is decreased by 1.

# Part 3 - The length and index of a string

The function len returns the number of characters in a string, which is always an integer value. For example, len("hey") returns 3, because there are three characters in the string hey.

The following program asks the user for a string and then prints it "underlined". The program prints a second line with as many - characters as is the length of the input:

input_string = input("Please type in a string: ")
print(input_string)
print("-"*len(input_string))

Please type in a string: Hi there!
Hi there!


As strings are essentially sequences of characters, any single character in a string can also be retrieved. The operator [] finds the character with the index specified within the brackets.

The index refers to a position in the string, counting up from zero. The first character in the string has index 0, the second character has index 1, and so forth.

For example, this program
input_string = input("Please type in a string: ")
print(input_string[0])
print(input_string[1])
print(input_string[3])

would print out this:
Please type in a string: monkey
m
o
k

Since the first character in a string has the index 0, the last character has the index length - 1. The following program prints out the first and the last characters of a string:

input_string = input("Please type in a string: ")
print("First character: " + input_string[0])
print("Last character: " + input_string[len(input_string) - 1])

Please type in a string: testing
First character: t
Last character: g


The following program loops through all the characters in a string from first to last:

input_string = input("Please type in a string: ")
index = 0
while index < len(input_string):
    print(input_string[index])
    index += 1

Please type in a string: test
t
e
s
t

You can also use negative indexing to access characters counting from the end of the string. The last character in a string is at index -1, the second to last character is at index -2, and so forth. You can think of input_string[-1] as shorthand for input_string[len(input_string) - 1].

The example from above can be simplified with negative indexing:
input_string = input("Please type in a string: ")
print("First character: " + input_string[0])
print("Last character: " + input_string[-1])

Please type in a string: testing
First character: t
Last character: g

* IndexError: string index out of range

If you tried the above examples for yourself, you may already have come across the error message IndexError: string index out of range. This error appears if you try to access an index which is not present in the string.

input_string = input("Please type in a string: ")
print("The tenth character: " + input_string[9])

Please type in a string: introduction to programming
The tenth character: i

Please type in a string: python
Traceback (most recent call last):
File "", line 1, in 
IndexError: string index out of range

Sometimes an indexing error is caused by a bug in the code. For example, it is quite common to index too far when trying to access the last character in a string:
input_string = input("Please type in a string: ")
print("Last character: " + input_string[len(input_string)])

Since string indexing begins at zero, the last character is at index len(input_string) - 1, not at len(input_string).

There are situations where the program should prepare for errors caused by input from the user:
input_string = input("Please type in a string: ")
if len(input_string) > 0:
    print("First character: " + input_string[0])
else:
    print("The input string is empty. There is no first character.")


In the example above, if the programmer hadn't included a check for the length of the input string, a string of length zero would have caused an error. A string of length zero is also called an empty string, and here it would be achieved by just pressing Enter at the input prompt.

```python
# Programming Exercise :
# The longer string

'''
Please write a program which asks the user for two strings and then prints out whichever is the longer of the two - 
that is, whichever has the more characters. If the strings are of equal length, the program should print out 
"The strings are equally long".

Some examples of expected behaviour:
Please type in string 1: hey
Please type in string 2: hiya
hiya is longer

Please type in string 1: howdy doody
Please type in string 2: hola
howdy doody is longer

Please type in string 1: hey
Please type in string 2: bye
The strings are equally long
'''

# ALGORITMA
'''
1. Mulai
2. Program minta input string pertama ke user
    2.1 User input string pertama ke program
3. Program minta input string kedua ke user
    3.1 User input string kedua ke program
4. Program mengolah kedua input dari user
    4.1 Cek panjang string pertama dan kedua menggunakan len()
    4.2 If string pertama > string kedua, print(f"{string pertama} is longer)
        elif string pertama < string kedua, print(f"{string kedua} is longer)
        else, print("The strings are equally long)
5. Selesai
'''

stringPertama = input("Please type in string 1: ")
stringKedua = input("Please type in string 2: ")

panjangStringPertama = len(stringPertama)
panjangStringKedua = len(stringKedua)

if panjangStringPertama > panjangStringKedua:
    print(f"{stringPertama} is longer")
elif panjangStringPertama < panjangStringKedua:
    print(f"{stringKedua} is longer")
else:
    print("The strings are equally long")
```

```python
# Programming Exercise :
# End to beginning

'''
Please write a program which asks the user for a string. The program then prints out the input string in reversed order, 
from end to beginning. Each character should be on a separate line.

An example of expected behaviour:
Please type in a string: hiya
a
y
i
h
'''

# ALGORITMA
'''
1. Mulai
    1.1 definisikan index = -1
2. Program minta input string ke user -> kata = input("Please type in a string: ")
3. User ngasih input string ke program
4. Program mengolah inputan string dari user
    4.1 WHILE index < 0:
        print(kata[index])
        index -= 1
        if len(kata) - index == 0:
            break
5. Selesai

REVISI 
    4.1 WHILE index >= -len(kata):  -> Gausah pake if lagi kalo udah kaya gini, nanti ketika indexnya makin mengecil, suatu
    saat dia bakal kurang dari si -len(kata), yang menyebabkan kondisi while-nya ngga terpenuhi lagi (berhenti loopnya)
        print(kata[index])
        index -= 1
'''
index = -1
kata = input("Please type in a string: ")

while index >= -len(kata):
    print(kata[index])
    index -= 1
```

```python
# Programming Exercise :
# Right-aligned

'''
Please write a program which asks the user for a string and then prints it out 
so that exactly 20 characters are displayed. If the input is shorter than 20 characters, 
the beginning of the line is filled in with * characters.

You may assume the input string is at most 20 characters long.

Please type in a string: python
**************python

Please type in a string: alongerstring
*******alongerstring

Please type in a string: averyverylongstring
*averyverylongstring
'''

# ALGORITMA / PSEUDOCODE
'''
1. Mulai program Right-aligned
    1.1 definisikan string bintangFull (bintang sebanyak 20 karakter) 
    buat keperluan string finalnya nanti -> bintangFull = "********************"
2. Program minta input string ke user -> inputStringUser = input("Please type in a string: ")
3. Program mengolah inputStringUser
    3.1 Set kondisi program bakal jalan hanya JIKA inputStringUser-nya maksimal 20 karakter. 
            IF inputStringUser <= 20:
                panjangBintangPrint = len(bintangFull) - len(inputStringUser) -> buat ngeprint "*" sebanyak (20 karakter - karakter dari string inputan user)
                stringFinal = ("*" * panjangBintangPrint) + inputStringUser -> jadi nanti outputnya adalah bintang di depan sebanyak (20 karakter - karakter dari string inputan user)
                ditambah sama inputStringUser itu sendiri.
                print(stringFinal)
4. Selesai



REVISI 

IF inputStringUser <= 20: -> Jangan lupa tambahin len-nya, kan mau ngebandingin panjang karakter(int) sama 20(int)

jadinya IF len(inputStringUser <= 20: )
'''

bintangFull = "********************"
inputStringUser = input("Please type in a string: ")

if len(inputStringUser) <= 20:
    panjangBintangPrint = len(bintangFull) - len(inputStringUser)
    stringFinal = ("*" * panjangBintangPrint) + inputStringUser
    print(stringFinal)
```

```python
# Programming Exercise :
# A framed word

'''
Please write a program which asks the user for a string and then prints out a frame of * characters 
with the word in the centre. The width of the frame should be 30 characters. 
You may assume the input string will always fit inside the frame.

If the length of the input string is an odd number, you may print out the word in 
either of the two possible centre locations.

Word: testing
******************************
*          testing           *
******************************

Word: python
******************************
*           python           *
******************************
'''

# KOTRETAN KASAR
'''
berarti string yang di tengah maksimal 28 karakter karena di ujung kiri sama kanan ada satu bintang, jadi kan 30 - (1 *2)

misal stringnya 5 karakter:
    28 - 5 = 23 space kosong kanan + kiri -> bisa kanan 12, kiri 11 ATAU kanan 11, kiri 12
    print("******************************")
    print("*" + (" " * (23//2)) + string + (" " * ((23//2)+1)))
    print("******************************")

    
misal stringnya 4 karakter:
    28 - 4 = 24 space kosong kanan + kiri -> kanan 12, kiri 12
    print((" " * (24//2)) + string)

string = "haloo" -> String GANJIL
print("******************************")
print("*" + (" " * (23//2)) + string + (" " * ((23//2)+1) + "*"))
print("******************************")
OUTPUT:
******************************
*           haloo            *
******************************


string = "halo" -> String GENAP
print("******************************")
print("*" + (" " * (24//2)) + string + (" " * ((24//2)) + "*"))
print("******************************")
OUTPUT:
******************************
*            halo            *
******************************
'''

# ALGORITMA / PSEUDOCODE
'''
1. Mulai program A Framed Word
2. Definisikan beberapa variabel
    Karena kalo ngeprint itu per baris, maka framenya gabisa langsung dibentuk jadi kotak, harus per baris juga. MAKA
        2.1 Definisikan frame bintang atas -> frameAtas = "*" * 30
        2.2 Definisikan juga frame bintang bawah -> frameBawah = "*" * 30
        2.3 Barulah kita bentuk isi tengahnya
            2.3.1 Sisa space kosong di tengah buat nampilin string inputan user kan ada 28 karakter.  
                  Minta input user.  Berarti nanti input dari usernya harus <=28 biar masih muat.
            2.3.2 Abis itu kita bikin kondisional, kondisi pertama jika panjang karakter string inputan user genap,
                  Kondisi kedua jika panjang karakter string inputan user ganjil.
            2.3.3 Biar nanti ngeprintnya enak, gausah banyak notasi mtk dan angka, 
                  maka nanti buat variabel aja jarakKiri dan jarakKanan di kondisional pertama dan kedua
            2.3.4 Udah deh nanti tinggal diprint kaya gini -> print("*" + jarakKiri + string + jarakKanan + "*")
3. Selesai

REVISI

-> jangan lupa bikin variabel selisihSpaceDanString buat nentuin pembaginya
'''

frameAtas = frameBawah = "*" * 30
inputString = input("Word: ")
selisihSpaceDanString = 28 - len(inputString) 

if len(inputString) <= 28 and len(inputString) % 2 == 0:  #Kondisi 1 (JIKA panjang inputan string user GENAP)
    jarakKiri = jarakKanan = " " * (selisihSpaceDanString // 2) #Karena genap jadi space di kanan dan kirinya sama
    print(frameAtas)
    print("*" + jarakKiri + inputString + jarakKanan + "*")
    print(frameBawah)
    
elif len(inputString) <= 28 and len(inputString) % 2 != 0: #Kondisi 2 (JIKA panjang inputan string user GANJIL)
    jarakKiri = " " * (selisihSpaceDanString // 2) #Karena ganjil jadi spacenya bisa antara yang kiri kurang 1 dari yang di kanan atau sebaliknya; di sini pake yang di kiri kurang 1 dari yang di kanan aja
    jarakKanan = jarakKiri + " "
    print(frameAtas)
    print("*" + jarakKiri + inputString + jarakKanan + "*")
    print(frameBawah)
```
